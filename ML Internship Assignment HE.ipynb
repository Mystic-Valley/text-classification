```python
#Importing Required Libraries

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd 
import numpy as np
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
import pickle
from nltk.corpus import stopwords
documents = []

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
from catboost import CatBoostClassifier
from sklearn.tree import DecisionTreeClassifier, ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import accuracy_score, classification_report

from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
df = pd.read_csv("hm_train.csv")
df_test = pd.read_csv('hm_test.csv')
```

```python
X, y = df.cleaned_hm, df.predicted_category
```

```python
df.predicted_category.unique()
```

```python
df_test.loc[67]
```

```python
X = X.replace('\n', '', regex=True)
X = X.replace('\r', '', regex = True)
vectorizer = CountVectorizer(max_features=1500, min_df=5, max_df=0.7, stop_words=stopwords.words('english'))
X = vectorizer.fit_transform(X).toarray()
tfidfconverter = TfidfTransformer()
X = tfidfconverter.fit_transform(X).toarray()
```

```python

def print_line():
    print("----------------------------------------------------------")

#Classification

def train_clf(train_x, test_x, train_y, test_y, clf):
    res = clf.fit(train_x, train_y).score(test_x, test_y)*100
    return res
    
    
def ClassificationModel(X, y, test_size = 0.25, shuffle = True, normalize = False):
    if normalize == True:
        si = StandardScaler()
        X = si.fit_transform(X)
    train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=test_size, shuffle=shuffle)
    
    xtc = ExtraTreeClassifier()
    rfc = RandomForestClassifier()
    xgb = XGBClassifier()
    dtc = DecisionTreeClassifier()
    mlp = MLPClassifier()
    lgbm = LGBMClassifier()
    
    print("Accuracy of Extra Tree Classifier is {}%".format(train_clf(train_x, test_x, train_y, test_y, xtc)))
    print_line()
    print("Accuracy of Decision Tree Classifier is {}%".format(train_clf(train_x, test_x, train_y, test_y, dtc)))
    print_line()
    print("Accuracy of Random Forest Classifier is {}%".format(train_clf(train_x, test_x, train_y, test_y, rfc)))
    print_line()
    print("Accuracy of XGBoost Classifier is {}%".format(train_clf(train_x, test_x, train_y, test_y, xgb)))
    print_line()
    print("Accuracy of Multi Layer Perceptron Classifier is {}%".format(train_clf(train_x, test_x, train_y, test_y, mlp)))
    print_line()
    print("Accuracy of LGBM Classifier is {}%".format(train_clf(train_x, test_x, train_y, test_y, lgbm)))
    print_line()
```

```python
ClassificationModel(X, y)
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.25, shuffle=True)
```

```python
clf = MLPClassifier(solver = 'sgd', learning_rate = 'adaptive', learning_rate_init = 0.001, verbose = True)
clf.fit(train_x, train_y).score(test_x, test_y)
```

```python
clf.fit(X, y)
```

```python
Xt = df_test.cleaned_hm
Xt = Xt.replace('\n', '', regex=True)
Xt = Xt.replace('\r', '', regex = True)
Xt = vectorizer.transform(Xt).toarray()
Xt = tfidfconverter.transform(Xt).toarray()

res = clf.predict(Xt)
```

```python
submission = pd.DataFrame({'hmid':df_test['hmid'],'predicted_category': res})
submission.to_csv('Submit.csv', index = False)
```
